```python
import pandas as pd
import sqlite3

# Load the data
file_path = '/content/Dataset for Data Analytics (2).xlsx'
df = pd.read_excel(file_path)

# Create SQLite database
conn = sqlite3.connect('ecommerce_analysis.db')

# Load dataframe to SQL table
df.to_sql('orders', conn, if_exists='replace', index=False)

print("Database created successfully!")
print(f" Loaded {len(df)} rows into 'orders' table")

# Verify
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(f"Tables: {cursor.fetchall()}")
```

```python
# View all columns and first 10 rows
# The 'conn' object is available from the previous cell.
query = "SELECT * FROM orders LIMIT 10;"

# Execute the query and fetch results into a pandas DataFrame for better display
df_result = pd.read_sql_query(query, conn)
print(df_result)
```

```python
# Select only relevant columns for analysis
query_filtered = """
SELECT
    OrderID,
    Date,
    Product,
    Quantity,
    UnitPrice,
    TotalPrice,
    OrderStatus
FROM orders
LIMIT 10;
"""

# Execute the query and fetch results into a pandas DataFrame
df_filtered_orders = pd.read_sql_query(query_filtered, conn)
print(df_filtered_orders)
```

```python
#Find all delivered orders
query_delivered_orders = """
SELECT
    OrderID,
    Product,
    TotalPrice,
    OrderStatus
FROM orders
WHERE OrderStatus = 'Delivered'
LIMIT 10;
"""

df_delivered_orders = pd.read_sql_query(query_delivered_orders, conn)
print(df_delivered_orders)
```

```python
# Find orders over $2000
query_high_value_orders = """
SELECT
    OrderID,
    Product,
    TotalPrice,
    OrderStatus
FROM orders
WHERE TotalPrice > 2000
ORDER BY TotalPrice DESC;
"""

df_high_value_orders = pd.read_sql_query(query_high_value_orders, conn)
print(df_high_value_orders)
```

```python
query_orders_2024 = """
SELECT
    OrderID,
    Date,
    Product,
    TotalPrice
FROM orders
WHERE Date BETWEEN '2024-01-01' AND '2024-12-31'
ORDER BY Date
LIMIT 10;
"""

df_orders_2024 = pd.read_sql_query(query_orders_2024, conn)
print(df_orders_2024)
```

```python
query_high_value_completed_orders = """
SELECT
    OrderID,
    Product,
    TotalPrice,
    OrderStatus,
    PaymentMethod
FROM orders
WHERE TotalPrice > 1000
    AND OrderStatus IN ('Delivered', 'Shipped')
ORDER BY TotalPrice DESC
LIMIT 10;
"""

df_high_value_completed_orders = pd.read_sql_query(query_high_value_completed_orders, conn)
print(df_high_value_completed_orders)
```

```python
query_product_counts = """
SELECT
    Product,
    COUNT(*) as OrderCount
FROM orders
WHERE Product LIKE '%Phone%'
    OR Product LIKE '%Laptop%'
GROUP BY Product;
"""

df_product_counts = pd.read_sql_query(query_product_counts, conn)
print(df_product_counts)
```

```python
query_highest_revenue_orders = """
SELECT
    OrderID,
    Product,
    TotalPrice,
    OrderStatus
FROM orders
ORDER BY TotalPrice DESC
LIMIT 10;
"""

df_highest_revenue_orders = pd.read_sql_query(query_highest_revenue_orders, conn)
print(df_highest_revenue_orders)
```

```python
query_sorted_orders = """
SELECT
    OrderStatus,
    Product,
    TotalPrice
FROM orders
ORDER BY OrderStatus, TotalPrice DESC
LIMIT 20;
"""

df_sorted_orders = pd.read_sql_query(query_sorted_orders, conn)
print(df_sorted_orders)
```

```python
query_product_metrics = """
SELECT
    Product,
    COUNT(*) as NumberOfOrders,
    SUM(Quantity) as TotalUnitsSold
FROM orders
GROUP BY Product
ORDER BY NumberOfOrders DESC
LIMIT 10;
"""

df_product_metrics = pd.read_sql_query(query_product_metrics, conn)
print(df_product_metrics)
```

```python
query_revenue_per_product = """
SELECT
    Product,
    COUNT(*) as OrderCount,
    SUM(TotalPrice) as TotalRevenue,
    AVG(TotalPrice) as AvgOrderValue,
    MIN(TotalPrice) as MinOrderValue,
    MAX(TotalPrice) as MaxOrderValue
FROM orders
GROUP BY Product
ORDER BY TotalRevenue DESC
LIMIT 10;
"""

df_revenue_per_product = pd.read_sql_query(query_revenue_per_product, conn)
print(df_revenue_per_product)
```

```python
query_monthly_analysis = """
SELECT
    strftime('%Y-%m', Date) as Month,
    COUNT(*) as OrderCount,
    SUM(TotalPrice) as TotalRevenue,
    AVG(TotalPrice) as AvgOrderValue
FROM orders
GROUP BY Month
ORDER BY Month;
"""

df_monthly_analysis = pd.read_sql_query(query_monthly_analysis, conn)
print(df_monthly_analysis)
```
